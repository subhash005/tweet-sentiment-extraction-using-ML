# Accuracy Score

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

x = np.array(df["tweet"])
y = np.array(df["category"])

cv = CountVectorizer()
x = cv.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)


classifiers = {
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Logistic Regression": LogisticRegression()
}

for name, clf in classifiers.items(): 
   
    clf.fit(x_train, y_train)
    
    y_pred = clf.predict(x_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    
    print(f"{name} Accuracy: {accuracy:.4f}")
```

```python
import tkinter as tk
from tkinter import ttk, messagebox
```

```python
import pandas as pd
import numpy as np
import re
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from nltk.corpus import stopwords
import tkinter as tk
from tkinter import messagebox
from tkinter import ttk

# Load the dataset
df = pd.read_csv('twitter.csv')

# Define a function to remove special characters
def remove_special_characters(text):
    pattern = r'[@!#\$%\^&\*\(\)_\+\=\{\}\[\]:;<>\?\'\"\\\|`~]'
    clean_text = re.sub(pattern, '', text)
    return clean_text

# Apply the function to the "tweet" column
df["tweet"] = df["tweet"].apply(remove_special_characters)

# Map labels to categories
df['category'] = df['label'].map(
    {0: "Good sentiment detected", 
     1: "Bad sentiment detected",
     2: "null"}
)

# Define features and labels
x = np.array(df["tweet"])
y = np.array(df["category"])

# Initialize CountVectorizer
cv = CountVectorizer(stop_words=stopwords.words("english"))

# Transform the text data to feature vectors
x = cv.fit_transform(x)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# Train the Decision Tree Classifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)

# Define the prediction function
def predict_tweet_category(tweet):
    transformed_tweet = cv.transform([tweet]).toarray()
    prediction = clf.predict(transformed_tweet)
    return prediction[0]

# Create the GUI
def on_predict():
    predict_button.config(bg="black")  # Set the button color to black before prediction
    tweet = combobox.get()
    if not tweet:
        messagebox.showwarning("Selection Error", "Please select a tweet from the dropdown.")
        return
    prediction = predict_tweet_category(tweet)
    
    # Change window background color based on prediction
    if prediction == "Good sentiment detected":
        root.config(bg="green")
    elif prediction == "Bad sentiment detected":
        root.config(bg="red")
    else:
        root.config(bg="gray")
    
    # Set result text color to white permanently
    result_label.config(text=f"ANALYSIS RESULT: {prediction}", fg="white")
    
    predict_button.config(bg="white")  # Set the button color to white after prediction

def on_clear():
    combobox.set('')
    result_label.config(text="ANALYSIS RESULT: ", fg="white")  # Keep the result text color white permanently
    root.config(bg='white')  # Reset to default background color
    predict_button.config(bg="black")  # Set the button color back to black when clearing

# Initialize the main window
root = tk.Tk()
root.title("Tweet Category Predictor")
root.geometry("800x400")

# Configure grid columns to be centered
root.grid_columnconfigure(0, weight=1)
root.grid_columnconfigure(1, weight=1)

# Create and place the widgets
tk.Label(root, text="Select Tweet:").grid(row=0, column=0, columnspan=2, padx=10, pady=10, sticky='ew')

# Create a Combobox to display tweets
combobox = ttk.Combobox(root, width=50)
combobox.grid(row=1, column=0, columnspan=2, padx=10, pady=10, sticky='ew')

# Add tweets to the Combobox
combobox['values'] = df["tweet"].tolist()

predict_button = tk.Button(root, text="ANALYSE", command=on_predict, fg="black", bg="black")
predict_button.grid(row=2, column=0, columnspan=2, pady=10, sticky='ew')

clear_button = tk.Button(root, text="Clear", command=on_clear)
clear_button.grid(row=3, column=0, columnspan=2, pady=10, sticky='ew')

result_label = tk.Label(root, text="ANALYSIS RESULT: ", bg=root.cget('bg'), fg="white")  # Set initial text color to white
result_label.grid(row=4, column=0, columnspan=2, pady=10, sticky='ew')

# Start the main loop
root.mainloop()
```
